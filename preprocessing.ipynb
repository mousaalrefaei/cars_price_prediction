```python
# import necessary libraries:
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# read data file:
df = pd.read_csv('cars.csv')
```

```python
# show the first 5 rows of the data:
df.head()
```

```python
# drop unnecessary column:
df = df.drop(columns=['url','country','date','Hubraum','CO₂-Emissionen','Kraftstoffverbrauch','Lackierung','Innenausstattung','Farbe laut Hersteller','Farbe der Innenausstattung','Umweltplakette','Schadstoffklasse','verkaufer','Türen','Sitzplätze'])
```

```python
# drop rented cars (which have 'mtl' with price)
df = df[~df['price'].str.contains('mtl')]

# drop duplicated headers: 
df = df[~df['price'].str.contains('price')]

# remove 'DE' from the ort values:
df['ort'] = df['ort'].str.replace(', DE','')

# remove new cars:
df = df[~df['Zustand'].str.contains('Neu')]
df = df.drop(columns='Zustand')

# drop duplicated rows:
df = df.drop_duplicates()
```

```python
# convert numeric columns to float:

df['price'] = df['price'].astype('float')
df['Kilometerstand'] = df['Kilometerstand'].str.replace('.','').str.replace(' km','').astype('float')

# df['Türen'] = df['Türen'].astype('float')
# df['Sitzplätze'] = df['Sitzplätze'].astype('float')

# keep just the kw in Leistung column and convert it to float:

df['Leistung'] = df['Leistung'].str.extract('(^\d+)').astype('float')


# remove the month from the  Erstzulassung column and convirt it to float:
df['Erstzulassung'] = df['Erstzulassung'].str.extract('(\d{4})').astype('float') 
```

```python
# outliers:

# we have just about 100 outliers in Kilometerstand column so will drop it :
df = df.loc[(df['Kilometerstand'] < 750000) | (np.isnan(df['Kilometerstand']))]

# drop the missing values in modell column:
df = df[df['modell'].notna()]
```

```python
# remove missing values:
df= df.dropna()
```

```python
# keep the important brands:

for x, i in enumerate(df['marke'].value_counts().tail(63).index):
  df= df[~df['marke'].str.contains(i)]
```
